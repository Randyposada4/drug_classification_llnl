## Data 
We utilize the Scalable Precision Medicine Open Knowledge Engine, which is described as a database of databases to create a machine learning model designed to classify small molecule drugs. Ultimately, we hope to build a model for general clustering or classification, but for now we use a subset of nodes and edges with labeled data. This labeled data is then used to train and test both the featurization and model design.

The graphical data is stored as nodes and edges with various labels, but for a baseline model we utilize the following labels:
- Compound
- Side Effect
- Disease
- Gene
- Protein
- Protein Domain

The descriptions of the above Nodes (and their corresponding properties) are available on the [SPOKE website](https://spoke.rbvi.ucsf.edu/docs/index.html).
